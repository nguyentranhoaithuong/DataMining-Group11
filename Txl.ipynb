```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
eps = np.finfo(float).eps
from numpy import log2 as log
from pprint import pprint
import seaborn as sns
# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv('diabetes.csv')
```

```python
# Vẽ đồ thị phân phối của dữ liệu gốc
sns.boxplot(df['Glucose'])
plt.title('Phân phối dữ liệu gốc')
plt.xlabel('Glucose')
plt.show()
```

```python
z_scores = np.abs((df['Glucose'] - df['Glucose'].mean()) / df['Glucose'].std())
threshold = 3
outlier_indices = z_scores[z_scores > threshold].index
# Loại bỏ các dòng có index là outlier_indices
df = df.drop(outlier_indices)
# Vẽ lại đồ thị sau khi loại bỏ outlier
sns.boxplot(data=df['Glucose'])
plt.title('Phân phối dữ liệu sau khi loại bỏ outlier')
plt.xlabel('Value')
plt.show()
```
