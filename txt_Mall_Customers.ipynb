```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
eps = np.finfo(float).eps
from numpy import log2 as log
from pprint import pprint
import seaborn as sns
# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv('Mall_Customers.csv')
```

```python
# Vẽ đồ thị phân phối của dữ liệu gốc
sns.boxplot(df['Annual Income (k$)'])
plt.title('Phân phối dữ liệu gốc')
plt.xlabel('Annual Income (k$)')
plt.show()
```

```python
z_scores = np.abs((df['Annual Income (k$)'] - df['Annual Income (k$)'].mean()) / df['Annual Income (k$)'].std())
threshold = 3
outlier_indices = z_scores[z_scores > threshold].index
# Loại bỏ các dòng có index là outlier_indices
df = df.drop(outlier_indices)
# Vẽ lại đồ thị sau khi loại bỏ outlier
sns.boxplot(data=df['Annual Income (k$)'])
plt.title('Phân phối dữ liệu sau khi loại bỏ outlier')
plt.xlabel('Value')
plt.show()
```

```python
# Vẽ đồ thị phân phối của dữ liệu gốc
sns.boxplot(df['Spending Score (1-100)'])
plt.title('Phân phối dữ liệu gốc')
plt.xlabel('Spending Score (1-100)')
plt.show()
```

```python
z_scores = np.abs((df['Spending Score (1-100)'] - df['Spending Score (1-100)'].mean()) / df['Spending Score (1-100)'].std())
threshold = 3
outlier_indices = z_scores[z_scores > threshold].index
# Loại bỏ các dòng có index là outlier_indices
df = df.drop(outlier_indices)
# Vẽ lại đồ thị sau khi loại bỏ outlier
sns.boxplot(data=df['Spending Score (1-100)'])
plt.title('Phân phối dữ liệu sau khi loại bỏ outlier')
plt.xlabel('Value')
plt.show()
```
